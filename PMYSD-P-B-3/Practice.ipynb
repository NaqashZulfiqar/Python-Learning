Python **Variable**

```python
#Variables
x = 5 #integer
y = 5.5 #float
name = "AI" #String
is_active = True #Boolean

print(type(x))
print(type(y))
print(type(name))
print(type(is_active))

#Multiple Assignments
a,b,c = 10,20,30
print(a,b,c)

#Same value to multiple variables
i = j = k ="AI"
print(i,j,k)

#delete variable
p = 5
del p
#print(p) #NameError: name 'p' is not defined
```

Python Data **Types**

```python
#Python Data Types
# 1=> Numeric Type , 2=> Sequence Type, 3=> Set Type, 4=> Mapping Type, 5=> Boolean Type , 6=> Binary Type

# 1=> Numeric Type
    # 1=> Numbers , 2=> Float , 3=> Complex
a = 5
b = 6.5
c = 3 + 2j
print(type(a))
print(type(b))
print(type(c))
```

```python
# 2=> Sequence Type
    # 1=> String , 2=> List , 3=> Tuple
text = "Machine Learning"
print(type(text))
#List []
numbers = [1,2,3,4]
numbers.append(5) #adding new value
print(numbers)
#Tuple
coordinates =(10,20,30)
print(coordinates[0])
```

```python
# 3=> Set Types
    # 1=> Set , 2=> Frozen Set
my_set = {1,2,3,3,4}
print(my_set) #duplicates are removed

setname = frozenset(my_set) #Immutable
```

```python
# 4=> Mapping Type (Dictionary)
student = {"name":"asad","age":"24","course":"AI",
           "name1":"adil","age1":"23","course1":"Java"}
print(student["name1"])
```

```python
# 5=> Boolean Type
x = True
y = False
print(type(x))
```

```python
# 6=> Binary Types
    # 1=> Byte (immutable) , 2=> ByteArray (mutable) , 3=> memoryview
byte_data = b"AI"
print(type(byte_data))

mutableBytes= bytearray([65,66,67])
mutableBytes[0] = 68
print(mutableBytes)
```
