```python
import numpy as np
import time

#Using Python List
size = 1_000_000 #1 million
py_list = list(range(size))
start_time = time.time()
sum_py = sum(py_list)
end_time = time.time()
print("Sm using Python list : ", sum_py)
print("Time Taken :",end_time - start_time,"seconds")

#Using Numpy array
np_array = np.arange(size)

start_time = time.time()
sum_py = np.sum(np_array)
end_time = time.time()

print("\nSum using NumPy array: ",sum_py)
print("Time Taken : ", end_time - start_time,"Seconds")
```

2. Memory Efficiency: NumPy Uses Less Memory

```python
import sys

#Python List memory usage
py_list = list(range(1000))
print("Python list size:",sys.getsizeof(py_list)* len(py_list),"bytes")

#NumPy array memory usage
np_array = np.arange(1000)
print("NumPy array ize:",np_array.nbytes,"bytes")
```

3. Vectorized Operations: NumPy is More Efficient

```python
py_list = [1,2,3,4,5]
py_result = [X * 2 for X in py_list]
print(py_result)
```

```python
np_array = np.array([1,2,3,4,5])
np_result = np_array * 2
print(np_result)
```
