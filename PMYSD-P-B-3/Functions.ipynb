```python
#functions
def greet(name):
    print(f"Hello,{name}! Welcome to AI & ML") #f is format specifier here used for name!
greet("Ajmal")
```

```python
#Function
def square(num):
    return num * num
result = square(5)
print("Square of 5 :", result)
```

```python
#Function Adding
def addition(a,b):
    return a + b
print("Sum : ",addition(10,20))
```

```python
marks = int(input("Please enter Student Marks :"))
def studenMarks(marks):
    if(marks >= 60 and marks <= 64):
        print("Student Grade is : C")
    elif(marks >= 65 and marks <=69):
        print("Student Grade is : C+")
    elif(marks >= 70 and marks <=74):
        print("Student Grade is : B")
    elif(marks >= 75 and marks <=79):
        print("Student Grade is : B+")
    elif(marks >= 80 and marks <=100):
        print("Student Grade is : A")
    else:
        print("You are Fail")
studenMarks (marks)
```
