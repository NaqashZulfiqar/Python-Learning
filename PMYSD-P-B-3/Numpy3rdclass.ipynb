4. NumPy Operations

```python
import numpy as np
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(arr1 **2)
```

1. Summation: np.sum()

```python
arr =np.array([[1,2,3],
              [4,5,6]])
print(np.sum(arr)) #sum of all element
print(np.sum(arr, axis=0)) #Solumn wise
print(np.sum(arr, axis=1))# Row wise

print(np.mean(arr)) #Mean of All Elements
```

3. Standard Deviation & Variance

```python
print(np.std(arr))
print(np.var(arr))
```

For 2D arrays:

```python
arr_2d = np.array([[30,10,50],
                   [20,40,60]])
print(np.sort(arr_2d,axis=0))
print(np.sort(arr_2d,axis=1))

print(np.min(arr_2d)) #Return value
print(np.argmin(arr_2d))# Return Index
```

5. Linear Algebra with NumPy

```python
A = np.array([[1,2],
              [3,4]])
B = np.array([[5,6],
              [7,8]])
result = np.dot(A,B) # do like paper work
print(result)
print("****")
r = A@B
print(r)
```

4. Inverse of a Matrix (np.linalg.inv())

```python
A = np.array([[4,7],
              [2,6]])
A_inv = np.linalg.inv(A)
print(A_inv)

print(A @ A_inv)

#point to be remenbered
# 1: matrix should not be
```

5. Eigenvalues and Eigenvectors (np.linalg.eig())

```python
A = np.array([[4,-2],
              [1,1]])
eignvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues :",eignvalues)
print("Eigenvectors : \n",eigenvectors)
```
