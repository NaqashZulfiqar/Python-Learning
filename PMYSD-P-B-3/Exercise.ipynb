3. Convert a given temperature from Celsius to Fahrenheit using the formula: F=(C×9/5)+32F = (C \times 9/5) + 32F=(C×9/5)+32

```python
celsius = float(input("Please enter temperature in Celsius: "))
def FahrenHeit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
fahrenheit = FahrenHeit(celsius)
print("Temperature in Fahrenheit:", fahrenheit)
```

12. Write a Python script that checks whether a given year is a leap year.

```python
year = int(input("Please enter year"))
def isLeapYear(year):
    if(year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(year,"is a leap year")
    else:
        print(year,"is not a leap year")

isLeapYear(year)
```

13. Develop a simple grading system:
    1.   Input: User enters marks (0-100).
    2.   Output: Print grades:
        1.   80-100: "A"
        2.   70-79: "B"
        3.   60-69: "C"
        4.   50- 60: "D"
        5.   <50: “F”





```python
marks = int(input("Please enter marks between 0-100 : "))
def grades(marks):
    if(marks > 80 and marks <= 100):
        print(f"Your marks is {marks} and Grade : A")
    elif(marks > 70 and marks <= 79):
        print(f"Your marks is {marks} and Grade : B")
    elif(marks > 60 and marks <= 69):
        print(f"Your marks is {marks} and Grade : C")
    elif(marks > 50 and marks <= 60):
        print(f"Your marks is {marks} and Grade : D")
    else:
        print(f"Your marks is {marks} and Grade : F")

grades(marks)
```

17. Print the Fibonacci series up to n terms (user input).

```python
n = int(input("Enter the number of terms : "))
def Fibonacci(n):
    a,b = 0,1
    if n < 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print("Fibonacci series:")
        print(a)
    else:
        print("Fibonacci series:")
        for i in range(n):
            print(a,end= " ")
            a,b = b,a + b

Fibonacci(n)
```
