```python
file_path = "lab9_texts.txt"

try:
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()
    print("Loaded text from:",file_path)
    print("Text length:",len(text))
    print("Preview:", text[:200],"...")
except FileNotFoundError:
    text = ""
    print("File not found. Please ensure", file_path, "exists in the current folder")
```

```python
text = (
    "This is NLP first lecture. We're sitting in the class and learning from Sir Asad Ehsan."
    "Its 2:00 O'clock right now and we have to finsih at 3:00 O'clock ..."
    "Artificial Intelligence is transforming many industries. Students practice tokenization and stop word removal using NLTK."
    "Clean text helps models focus on important words in analysis."
)

print("Loaded sample string.")
print("Text Length: ", len(text))
print("Preview:", text[:200],"...")
# Artificial Intelligence is transforming many industries. Students practice tokenization and stop word
# removal using NLTK. Clean text helps models focus on important words in analysis.
```

Task 3: Cleaing Text

```python
import re
# re stands for regular expression

clean_text = text.lower()
clean_text = re.sub(r"[^a-z\s]", " ",clean_text)
# ^ sign mean not in
# a-z mean from a to z
# \s mean single space
# \s+ mean more than one space
clean_text = re.sub(r"\s+", " ", clean_text).strip()

print("Cleaned text length :", len(clean_text))
print(clean_text[:200], "...")
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

#Run once
nltk.download('punkt_tab')
```

```python
text =  "AI is changing the world. Students are learning NLP. Thi is fun!"

sentences = sent_tokenize(text)
words = word_tokenize(text)

print("Sentence:", sentences)
print("Words:", words)
```

Task 5. Stop-word removal using nltk

```python
import nltk
from nltk.corpus import stopwords

#Run once
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w not in stop_words]
print("Filtered Words: ", filtered_words[:25])
```

Task 6. Basic Frequency Analysis

```python
from collections import Counter

freq = Counter(filtered_words)
#Show
top10 = freq.most_common(10)
print("Top 10 terms: ", top10)
```
