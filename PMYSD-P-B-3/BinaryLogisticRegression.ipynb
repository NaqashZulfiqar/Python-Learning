```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#Binary Logistic Regression

#Input
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)

#output Pass(1) / Fail(0)
y = np.array([0,0,0,0,1,1,1,1])

#Train model
model = LogisticRegression()
model.fit(X,y)

#Predction
prediction = model.predict([[5]])
print("Prediction (Pass=1, Fail=0)", prediction)

#probability
prob = model.predict_proba([[5]])
print("Probability :",prob)
```

Multi Regresion

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#Input feature : weight
X = np.array([
    [150],
    [170],
    [140],
    [130],
    [200],
    [220]
])

#Output
y = np.array([0,0,1,1,2,2])

#Train Model
model = LogisticRegression(multi_class='ovr')
model.fit(X,y)

#Prediction
prediction = model.predict([[160]])
print("Predicted Fruit Class :", prediction)
```
